```python
# Calcular area del Triangulo
# Obtener altura y guardarla en variable H
# Obtener Base y guardarla en B
# Obtener area con la formula A = (B*H)/2
print("programa para calcular el area de un triangulo")
H = float(input("ingrese la altura del triangulo: "))
B = float(input("ingrese la base del triangulo: "))
A = (B*H)/2
print("el area del triangulo es: ", A)
print("Programa realizado por Angela Michelle Correa Lopez NL = 13")
```
